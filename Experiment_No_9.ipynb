## **Theory**

### **Histogram**
A **Histogram** is a graphical representation of the distribution of a numerical variable.  
It divides the data into **bins** and shows how many values fall into each bin.

- **X-axis:** Data range (bins)  
- **Y-axis:** Frequency  

### **Purpose**
- Understand data distribution  
- Detect skewness  
- Find spread and outliers  
- Identify patterns in numerical data  

---

### **Scatter Plot**
A **Scatter Plot** shows the relationship between two numerical variables.  
Each point represents a single observation.

- **X-axis:** Independent variable  
- **Y-axis:** Dependent variable  

### **Purpose**
- Detect relationship between variables  
- Identify clusters  
- Observe trends & patterns  
- Detect correlation  

---


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
import pandas as pd

# ---------------------------------------
# Load Wine Dataset
# ---------------------------------------
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)

# ---------------------------------------
# Histogram
# ---------------------------------------
plt.figure(figsize=(8, 5))
plt.hist(df['alcohol'], bins=12, color='skyblue', edgecolor='black')
plt.title("Histogram - Alcohol Content", fontsize=14, fontweight="bold")
plt.xlabel("Alcohol")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()

# ---------------------------------------
# Scatter Plot
# ---------------------------------------
plt.figure(figsize=(8, 5))
plt.scatter(df['alcohol'], df['malic_acid'], c='green', s=50, alpha=0.7, edgecolors='black')
plt.title("Scatter Plot - Alcohol vs Malic Acid", fontsize=14, fontweight="bold")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.grid(alpha=0.4)
plt.show()
```
